```python
import pandas as pd

# Read the input CSV file starting from the second row
df = pd.read_csv('ip_1.csv', skiprows=1)

# Group by course_code and aggregate roll numbers into a list
grouped_df = df.groupby('course_code')['rollno'].apply(list).reset_index()

# Rename columns
grouped_df.columns = ['course_code', 'registered_roll_numbers']

# Save the result to a new CSV file
grouped_df.to_csv('output_pandas.csv', index=False)

print("The output CSV file has been created successfully using pandas.")
```

```python
import pandas as pd

# Read the input CSV files
df_ip1 = pd.read_csv('ip_1.csv', skiprows=1)
df_ip2 = pd.read_csv('ip_2.csv', skiprows=1)

# Extract the list of courses from ip_2 for both morning and evening
morning_courses = df_ip2['Morning'].str.split('; ').explode().unique()
evening_courses = df_ip2['Evening'].str.split('; ').explode().unique()

# Combine morning and evening courses into a single set
valid_courses = set(morning_courses).union(set(evening_courses))

# Filter the courses in ip_1 that are present in the valid_courses set
filtered_df = df_ip1[df_ip1['course_code'].isin(valid_courses)]

# Group by course_code and aggregate roll numbers into a list, also count the number of students
grouped_df = filtered_df.groupby('course_code').agg(
    registered_roll_numbers=('rollno', list),
    number_of_students=('rollno', 'count')
).reset_index()

# Save the result to a new CSV file
grouped_df.to_csv('filtered_output.csv', index=False)

print("The filtered output CSV file has been created successfully using pandas.")
```

```python
import pandas as pd
import os

# Directory where the output files will be stored
output_directory = 'student_count_room_wise'

# Create the directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

directory = 'exam_schedule_day_wise'

def room_alloc(i):
    rooms_df = pd.read_csv('sorted_ip_3.csv')
    file_path = os.path.join(directory, f'day{i}.csv')

    if os.path.exists(file_path):
        courses_df = pd.read_csv(file_path)
    else:
        print(f"File 'day{i}.csv' not found in the directory.")
        return

    room_capacity_dict = dict(zip(rooms_df['Room Number'], rooms_df['Capacity']))
    course_student_dict = dict(zip(courses_df['course_code'], courses_df['number_of_students']))

    output_data = []
    room_index = 0
    room_remaining_capacity = room_capacity_dict.copy()

    for course_code, num_students in course_student_dict.items():
        roll_list = eval(courses_df[courses_df['course_code'] == course_code]['registered_roll_numbers'].values[0])
        day_value = courses_df[courses_df['course_code'] == course_code]['day'].values[0]
        date_value = courses_df[courses_df['course_code'] == course_code]['date'].values[0]

        remaining_students = num_students
        allocated_rooms = []
        room_strengths = []
        student_roll_numbers = []

        while remaining_students > 0:
            room_number = list(room_capacity_dict.keys())[room_index]
            room_capacity = room_remaining_capacity[room_number]

            if remaining_students <= room_capacity:
                allocated_rooms.append(room_number)
                room_strengths.append(remaining_students)
                student_roll_numbers.append('; '.join(roll_list[:remaining_students]))
                room_remaining_capacity[room_number] -= remaining_students
                remaining_students = 0
            else:
                allocated_rooms.append(room_number)
                room_strengths.append(room_capacity)
                student_roll_numbers.append('; '.join(roll_list[:room_capacity]))
                roll_list = roll_list[room_capacity:]
                room_remaining_capacity[room_number] = 0
                remaining_students -= room_capacity

            if room_remaining_capacity[room_number] == 0:
                room_index = (room_index + 1) % len(room_capacity_dict)

        output_data.append([day_value, date_value, course_code, '; '.join(student_roll_numbers), '; '.join(allocated_rooms), '; '.join(map(str, room_strengths))])

    output_df = pd.DataFrame(output_data, columns=['day', 'date', 'course_code', 'student_roll_numbers', 'allocated_rooms', 'room_strengths'])

    # Construct the output file path within the 'student_count_room_wise' directory
    output_file_path = os.path.join(output_directory, f'output_f{i}.csv')
    output_df.to_csv(output_file_path, index=False)

    print(f"Filtered and sorted courses for day {i} have been saved to {output_file_path}.")


for i in range(1, 14, 1):
    room_alloc(i)

print('Student_count_per_room');
```

```python
import pandas as pd
import os

directory = 'exam_schedule_day_wise'
directory_2 = 'student_count_room_wise'
def room_allocation_by_room(input_1, output_1, i):
    rooms_df = pd.read_csv('sorted_ip_3.csv')
    # courses_df = pd.read_csv(f'day{i}.csv')
    file_path = os.path.join(directory, f'day{i}.csv')
    if os.path.exists(file_path):
        courses_df = pd.read_csv(file_path)
    else:
        print(f"File not found: {file_path}")
        return


    # df = pd.read_csv(f'output_f{i}.csv')
    file_path_2= os.path.join(directory_2, f'output_f{i}.csv')
    if os.path.exists(file_path_2):
        df = pd.read_csv(file_path_2)
    else:
        print(f"File not found: {file_path_2}")
        return

    course_roll_numbers_dict = {}
    for index, row in courses_df.iterrows():
        course_roll_numbers_dict[row['course_code']] = eval(row['registered_roll_numbers'])

    new_data = []

    for index, row in df.iterrows():
        room_numbers = row['allocated_rooms'].split('; ')
        room_strengths = row['room_strengths'].split('; ')
        student_roll_numbers = row['student_roll_numbers'].split('; ')
        course_code = row['course_code']
        day_value = row['day']
        date_value = row['date']

        l = [0]
        x = 0
        for j in range(len(room_strengths)):
            x = x + int(room_strengths[j])
            l.append(x)

        all_roll_numbers_for_course = course_roll_numbers_dict[course_code]

        for j, room_number in enumerate(room_numbers):
            current_room_roll_numbers = student_roll_numbers[j].split('; ')
            roll_numbers_for_room = all_roll_numbers_for_course[int(l[j]):int(l[j + 1])]

            new_data.append([
                day_value,
                date_value,
                room_number,
                course_code,
                roll_numbers_for_room,
                room_strengths[j]
            ])

    new_df = pd.DataFrame(new_data, columns=['day','date','Room Number', 'Course Code', 'Roll Numbers', 'Room Strength'])

    # Create the 'task_1_output' directory if it doesn't exist
    output_dir = 'task_1_output'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Save the new DataFrame to a CSV file inside 'task_1_output'
    output_path = os.path.join(output_dir, f'rday{i}.csv')
    new_df.to_csv(output_path, index=False)

    print(f"Room allocation by room has been saved to {output_path}.")


for i in range(1, 14, 1):
    room_allocation_by_room(f'day{i}', f'output_f{i}', i)
```

```python
import csv
import os

# Create a new folder named "attendance sheet"
os.makedirs("attendance_sheet_output_2", exist_ok=True)
directory = 'task_1_output'
def attendance_sheet_day_wise(i):
    # Load the roll number to name mapping from ip_4.csv
    roll_name_mapping = {}
    with open('ip_4.csv', mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            roll_name_mapping[row['Roll']] = row['Name']

    # Load the room-wise data from rday1.csv
    room_data = {}
    file_path = os.path.join(directory, f'rday{i}.csv')
    if os.path.exists(file_path):
        courses_df = pd.read_csv(file_path)
    else:
        print(f"File not found: {file_path}")
        return

    with open(file_path, mode='r') as file: # Fix: use f-string to dynamically generate file name
        reader = csv.DictReader(file)
        for row in reader:
            room_number = row['Room Number']
            roll_numbers = eval(row['Roll Numbers'])
            day = row['day']
            date = row['date']
            if room_number not in room_data:
                room_data[room_number] = {'day': day, 'date': date, 'rolls': []}
            for roll in roll_numbers:
                room_data[room_number]['rolls'].append(roll)

    # Create the attendance sheet
    attendance_sheet = []
    for room, data in room_data.items():
        day = data['day']
        date = data['date']
        for roll in data['rolls']:
            name = roll_name_mapping.get(roll, 'Unknown')
            attendance_sheet.append([day, date, room, roll, name, ''])

    # Add 5 rows for signatures of invigilator and TAs
    for _ in range(5):
        attendance_sheet.append(['', '', '', '', '', ''])

    # Add 5 rows and 2 columns for TA's Name and TA's Sign at the end of the student data
    attendance_sheet.append(['', '', '', '', '', ''])
    attendance_sheet.append(["TA's Name", 'Sign'])
    for _ in range(4):
        attendance_sheet.append(['', ''])

    # Write the attendance sheet to day1_attendance.csv in the "attendance sheet" folder
    output_file_path = os.path.join("attendance_sheet_output_2", f'day{i}_attendance.csv') # Fix: use f-string to dynamically generate file name
    with open(output_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Day', 'Date', 'Room Number', 'Roll Number', 'Name', 'Sign'])
        writer.writerows(attendance_sheet)

    print(f"Attendance sheet has been generated and saved as {output_file_path}.")

for i in range(1,14,1):
  attendance_sheet_day_wise(i)
```
